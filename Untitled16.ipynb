## Data Cleaning for Visualization

### Subtask:
Standardize 'Safety_Certified' and 'Returned' columns and filter out negative 'Unit_Price_INR' values to ensure accurate plotting.


**Reasoning**:
I will clean and standardize the dataset by filtering out negative prices, mapping inconsistent categorical values in 'Safety_Certified' and 'Returned', handling missing data, and converting 'Transaction_Date' to datetime for future analysis.



```python
import pandas as pd

# 1. Filter out negative Unit_Price_INR values
df = df[df['Unit_Price_INR'] >= 0].copy()

# 2. Standardize 'Safety_Certified' and 'Returned' columns
# Map inconsistent strings to a clean 'Yes'/'No' format
standard_map = {
    'TRUE': 'Yes', 'true': 'Yes', 'yes': 'Yes', 'Yes': 'Yes', '1': 'Yes',
    'FALSE': 'No', 'false': 'No', 'no': 'No', 'No': 'No', '0': 'No'
}

df['Safety_Certified'] = df['Safety_Certified'].astype(str).str.strip().map(standard_map)
df['Returned'] = df['Returned'].astype(str).str.strip().map(standard_map)

# 3. Handle missing values by filling with 'Unknown'
df['Safety_Certified'] = df['Safety_Certified'].fillna('Unknown')
df['Returned'] = df['Returned'].fillna('Unknown')

# 4. Convert Transaction_Date to datetime object using mixed format parsing
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], dayfirst=True, format='mixed')

# 5. Verification
print(f"Rows remaining after filtering negative prices: {len(df)}")
print(f"Minimum Unit Price: {df['Unit_Price_INR'].min()}")
print(f"Unique Safety_Certified: {df['Safety_Certified'].unique()}")
print(f"Unique Returned: {df['Returned'].unique()}")
print(f"Transaction_Date dtype: {df['Transaction_Date'].dtype}")
display(df.head())
```

## Generate Multi-Chart Visualization

### Subtask:
Create a figure with multiple subplots: a bar chart for sales by 'Product_Category', a pie chart for 'Brand_Type' distribution, a scatter plot of 'Unit_Price_INR' vs 'Net_Revenue_INR', a line chart for 'Net_Revenue_INR' trends over years, and a stair plot for monthly quantity sold.


**Reasoning**:
I will import the necessary visualization libraries and create a multi-subplot figure as specified, including the bar, pie, scatter, line, and stair plots to visualize the Indian toy market sales data.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set consistent style
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(3, 2, figsize=(18, 18))

# 1. Bar Chart: Sales by Product Category
cat_sales = df.groupby("Product_Category")["Net_Revenue_INR"].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=cat_sales, x="Net_Revenue_INR", y="Product_Category", ax=axes[0, 0], palette="viridis", hue="Product_Category", legend=False)
axes[0, 0].set_title("Total Net Revenue by Product Category")
axes[0, 0].set_xlabel("Net Revenue (INR)")

# 2. Pie Chart: Brand Type Distribution
brand_counts = df["Brand_Type"].value_counts()
axes[0, 1].pie(brand_counts, labels=brand_counts.index, autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
axes[0, 1].set_title("Distribution of Brand Types")

# 3. Scatter Plot: Unit Price vs Net Revenue
sns.scatterplot(data=df, x="Unit_Price_INR", y="Net_Revenue_INR", alpha=0.5, ax=axes[1, 0], color="coral")
axes[1, 0].set_title("Unit Price vs Net Revenue Correlation")
axes[1, 0].set_xlabel("Unit Price (INR)")
axes[1, 0].set_ylabel("Net Revenue (INR)")

# 4. Line Chart: Net Revenue Trends over Years
yearly_revenue = df.groupby("Year")["Net_Revenue_INR"].sum().reset_index()
sns.lineplot(data=yearly_revenue, x="Year", y="Net_Revenue_INR", marker="o", ax=axes[1, 1], color="teal", linewidth=2.5)
axes[1, 1].set_title("Annual Net Revenue Trends (2020-2024)")
axes[1, 1].set_xticks(yearly_revenue["Year"])
axes[1, 1].set_ylabel("Total Net Revenue (INR)")

# 5. Stair Plot: Monthly Quantity Sold
monthly_qty = df.groupby("Month")["Quantity"].sum().reset_index()
axes[2, 0].step(monthly_qty["Month"], monthly_qty["Quantity"], where="mid", color="purple", linewidth=2)
axes[2, 0].fill_between(monthly_qty["Month"], monthly_qty["Quantity"], step="mid", alpha=0.2, color="purple")
axes[2, 0].set_title("Monthly Total Quantity Sold (Stair Plot)")
axes[2, 0].set_xlabel("Month")
axes[2, 0].set_ylabel("Total Quantity")
axes[2, 0].set_xticks(range(1, 13))

# Hide the empty 6th subplot
axes[2, 1].axis("off")

plt.tight_layout()
plt.show()
```
